```python
# import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the file into dataframe

ev_df = pd.read_csv('dataset/Electric_Vehicle_Population_Data.csv')
```

##### Data Cleaning

```python
ev_df.head()
```

```python
ev_df.sort_values(by='Electric Range', ascending=True).head()
```

```python
# Drop the columns 'Electric Range', 'Base MSRP, '2020 Census Tract', and 'Legislative District'.
# Drop 'state' also as we know the data is for Washington State (WA).

ev_df = ev_df.drop(columns=['State','Electric Range','Base MSRP','2020 Census Tract','Legislative District'])
ev_df.head()
```

```python
# Rewrite the 'Clean Alternative Fuel Vehicle (CAFV) Eligibility' values shorter
ev_df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] = ev_df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].str.replace('Eligibility unknown as battery range has not been researched', 'Unknown')
ev_df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] = ev_df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].str.replace('Clean Alternative Fuel Vehicle Eligible', 'Eligible')
ev_df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] = ev_df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].str.replace('Not eligible due to low battery range', 'Ineligible')
```

```python
ev_df.head()
```

```python
# Check if there are any duplicates

print(ev_df.duplicated().sum())
```

```python
# Determine missing values

print(ev_df.isna().sum())
```

```python
missing_county = ev_df[ev_df['County'].isna()]
missing_county
```

```python
# Drop rows with NaN. Since all six columns have missing values in the same rows, we can specify one column to drop the rows
ev_df = ev_df.dropna(subset=['County'])
```

```python
ev_df.isna().sum()
```

```python
# Remove any leading and trailing whitespaces from the columns with strings

columns = ['County','City','Make','Model','Electric Vehicle Type','Clean Alternative Fuel Vehicle (CAFV) Eligibility','Vehicle Location','Electric Utility']

for col in columns:
    ev_df[col] = ev_df[col].str.strip()
```

```python
# Export to CSV for further analysis using Excel and Power BI
ev_df.to_csv('dataset/ev_data_population_cleaned.csv')
```

#### Exploratory Data Analysis

```python
# Total current EV in WA state 

total_ev = ev_df['Electric Vehicle Type'].count()
print(f'Current total EVs in WA: {total_ev}')
```

```python
#The total number of BEV and PHEV in WA state

total_bev_phev = ev_df['Electric Vehicle Type'].value_counts().reset_index()
total_bev_phev.columns = ['EV Type', 'Total Count']
total_bev_phev
```

```python
COLORS = [
    '#FF5733',
    '#33B5FF',
    '#FF33A1',
    '#33FF57',
    '#FFBD33',
    '#8D33FF',
    '#33FFF2',
    '#FF3333',
    '#74FF33',
    '#3377FF'
]
ax = total_bev_phev.plot.bar(
    x='EV Type', 
    y='Total Count', 
    legend=False, 
    color=COLORS,
    figsize=(8, 6)
)

# Title and labels
plt.title('Total Number of BEV and PHEV in Washington State', fontsize=14, pad=20)
plt.xlabel('Type of Electric Vehicle', fontsize=12, labelpad=10)
plt.ylabel('Total Count', fontsize=12, labelpad=10)

# Formatting
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set proper y-axis scale
plt.ylim(0, 220000) 

# Add value labels on top of bars
for p in ax.patches:
    ax.annotate(f"{p.get_height():,}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points',
                fontsize=8
                )

# Improve layout
plt.tight_layout()
plt.show()
```

```python
# City in WA with the most EV

most_ev_city = ev_df['City'].value_counts().reset_index()
most_ev_city.columns = ['City','EV Count']
most_ev_city.head(10)
```

```python
ax = most_ev_city.head(10).plot.bar(
    x='City',
    y='EV Count',
    legend = False,
    color = COLORS, # type: ignore
    figsize = (8,6)
    )

plt.title('Most EV by a City in WA', fontweight='bold', pad=20)
plt.xlabel('City', labelpad=10)
plt.ylabel('EV Count', labelpad=10)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 42000)

for p in ax.patches:
    ax.annotate(f"{p.get_height():,}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points',
                fontsize=8)
    
plt.tight_layout()
plt.show()
```

```python
# Most sold EV make models in WA

top10_most_make_models = ev_df.groupby(['Make', 'Model']).size().reset_index(name='Total Count').sort_values(by='Total Count', ascending=False).head(10)
top10_most_make_models
```

```python
# Most Sold EV Make Models Sold in Washington State

top10_most_make_models['Make_Model'] = top10_most_make_models['Make'] + ' ' + top10_most_make_models['Model'] # Creates another column with make and model names
ax = top10_most_make_models.plot.bar(x='Make_Model',
                                     y='Total Count',
                                     legend=False,
                                     color=COLORS, # type: ignore
                                     figsize=(8,6)
                                     )
plt.title('Most Sold EV Make Models Sold in WA', fontweight='bold', pad=20)
plt.xlabel('Make Model', labelpad=10)
plt.ylabel('Total Count', labelpad=10)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 60000)

for p in ax.patches:
    ax.annotate(f"{p.get_height():,}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points',
                fontsize=8
                )
    
plt.tight_layout()
plt.show()
```

```python
# Filter out vehicles eligibility for Clean Alternative Fuel by VIN and export to CSV

eligibility_categories = {
    'Unknown': 'unknown_clean_alternative_fuel_eligibility.csv',
    'Eligible': 'clean_alternative_fuel_eligible.csv',
    'Ineligible': 'clean_alternative_fuel_ineligible.csv'
}

# Columns to select for output
cols = ['VIN (1-10)', 'DOL Vehicle ID', 'Make', 'Model', 'Model Year']

# Directory to save CSVs
output_dir = 'dataset/filtered_clean_fuel_eligibility/'

for status, filename in eligibility_categories.items():
    filtered = ev_df[ev_df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == status][cols]
    filtered.to_csv(output_dir + filename, index=False)
```
